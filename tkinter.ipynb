```python
import tkinter as tk
from tkinter import messagebox, ttk
import sqlite3

# Create/connect to SQLite database
conn = sqlite3.connect('students.db')
user = conn.cursor()

# Create table if it doesn't exist
user.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        age INTEGER,
        gender TEXT,
        email TEXT,
        mobile TEXT
    )
''')
conn.commit()
# Initialize main window
win = tk.Tk()
win.geometry("500x500")
win.title("Student Registration Portal")
def Registration():
    name = e1.get()
    age = e2.get()
    gender = e3.get()
    email = e4.get()
    mobile = e5.get()

    if not name or not age:
        messagebox.showerror("Input Error", "Please enter at least Name and Age.")
        return
    try:
        user.execute('''
            INSERT INTO students (name, age, gender, email, mobile)
            VALUES (?, ?, ?, ?, ?)
        ''', (name, age, gender, email, mobile))
        conn.commit()
        messagebox.showinfo("Success", "Student registered successfully!")
        e1.delete(0, tk.END)
        e2.delete(0, tk.END)
        e3.delete(0, tk.END)
        e4.delete(0, tk.END)
        e5.delete(0, tk.END)
    except Exception as e:
        messagebox.showerror("Database Error", f"An error occurred: {e}")
# âœ… Function to view all students in table
def View_Records():
    view_win = tk.Toplevel(win)
    view_win.title("Stored Student Records")
    view_win.geometry("700x400")

    # Create Treeview (table)
    cols = ("ID", "Name", "Age", "Gender", "Email", "Mobile")
    tree = ttk.Treeview(view_win, columns=cols, show="headings")

    # Define column headings
    for col in cols:
        tree.heading(col, text=col)
        tree.column(col, anchor="center", width=100)

    tree.pack(fill="both", expand=True)

    # Fetch records from DB
    user.execute("SELECT * FROM students")
    rows = user.fetchall()

    # Insert records into the table
    for row in rows:
        tree.insert("", tk.END, values=row)

    if not rows:
        messagebox.showinfo("Info", "No student records found!")

# Labels
l1 = tk.Label(win, text="Student details", font=("Arial", 16))
l2 = tk.Label(win, text="Name")
l3 = tk.Label(win, text="Age")
l4 = tk.Label(win, text="Gender")
l5 = tk.Label(win, text="Email")
l6 = tk.Label(win, text="Mobile number")

l1.grid(row=1, column=1, columnspan=2, pady=10)
l2.grid(row=2, column=1, sticky='e', padx=5, pady=5)
l3.grid(row=3, column=1, sticky='e', padx=5, pady=5)
l4.grid(row=4, column=1, sticky='e', padx=5, pady=5)
l5.grid(row=5, column=1, sticky='e', padx=5, pady=5)
l6.grid(row=6, column=1, sticky='e', padx=5, pady=5)

# Entry fields
e1 = tk.Entry(win)
e2 = tk.Entry(win)
e3 = tk.Entry(win)
e4 = tk.Entry(win)
e5 = tk.Entry(win)

e1.grid(row=2, column=2, padx=5, pady=5)
e2.grid(row=3, column=2, padx=5, pady=5)
e3.grid(row=4, column=2, padx=5, pady=5)
e4.grid(row=5, column=2, padx=5, pady=5)
e5.grid(row=6, column=2, padx=5, pady=5)

# Buttons
b1 = tk.Button(win, text="Submit Here", command=Registration)
b1.grid(row=7, column=1, columnspan=2, pady=10)

b2 = tk.Button(win, text="View Records", command=View_Records)
b2.grid(row=8, column=1, columnspan=2, pady=10)

win.mainloop()

# Close database when program exits
conn.close()
```
